Gr√°fico sugerido para ilustrar a distribui√ß√£o normal:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definindo a m√©dia e o desvio padr√£o
mu = 20  # m√©dia em mm
sigma = 5  # desvio padr√£o em mm

# Criando um intervalo de valores para a quantidade de chuva
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
# Calculando a fun√ß√£o de densidade de probabilidade (PDF)
pdf = norm.pdf(x, mu, sigma)

# Plotando a distribui√ß√£o normal
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribui√ß√£o Normal', color='blue')
plt.title('Distribui√ß√£o Normal da Quantidade de Chuva em Dias Chuvosos')
plt.xlabel('Quantidade de Chuva (mm)')
plt.ylabel('Densidade de Probabilidade')
plt.axvline(mu, color='red', linestyle='--', label='M√©dia (20 mm)')
plt.legend()
plt.grid()
plt.show()

# Exemplo: calcular a probabilidade de um dia ter menos de 25 mm de chuva
prob_menos_25 = norm.cdf(25, mu, sigma)
print(f"Probabilidade de um dia chuvoso ter menos de 25 mm de chuva: {prob_menos_25:.4f}")
```

3. Probabilidade Condicional: Rela√ß√£o entre Temperatura e Chuva
Aplicamos o conceito de probabilidade condicional para analisar como a temperatura influencia a quantidade de chuva em um dia. Sabemos que a correla√ß√£o (
ùúå
œÅ) entre a quantidade de chuva e a temperatura √© de -0.5, indicando uma correla√ß√£o negativa.

A f√≥rmula da distribui√ß√£o conjunta normal bivariada foi utilizada para calcular a probabilidade condicional entre a chuva e a temperatura. (Aqui, descrevemos o c√°lculo que voc√™ fez anteriormente).

```python
from scipy.stats import norm

# Definindo os par√¢metros para a temperatura
mu_T = 25  # m√©dia em ¬∞C
sigma_T = 4  # desvio padr√£o em ¬∞C

# Definindo o limite da temperatura
limite_temperatura = 24

# Calculando a probabilidade da temperatura ser menor que 24¬∞C
P_T_menor_24 = norm.cdf(limite_temperatura, mu_T, sigma_T)

# Exibindo a probabilidade
print(f"Probabilidade de a temperatura ser menor que {limite_temperatura}¬∞C: {P_T_menor_24:.4f}")
```

4. Probabilidade de Alagamentos e Custo Associado
Analisamos a frequ√™ncia dos alagamentos nos dias chuvosos e o impacto financeiro deles. Sabemos que cada alagamento tem um custo m√©dio de R$ 10.000.

Se a probabilidade de alagamentos nos dias chuvosos for calculada como
ùëÉ
(

)
P(Alagamento‚à£Chuva), podemos estimar os custos m√©dios mensais associados ao alagamento.

(Descrever o c√°lculo que fizemos e seus resultados).

```python
from scipy.stats import poisson

# Definindo a m√©dia de alagamentos por m√™s
lambda_alagamentos = 10  # m√©dia de alagamentos
custo_reparo = 10000  # custo por alagamento

# N√∫mero de alagamentos que queremos calcular a probabilidade
k = 10

# Calculando a probabilidade de ocorrer exatamente k alagamentos
probabilidade_k_alagamentos = poisson.pmf(k, lambda_alagamentos)

# Calculando o custo total se ocorrerem k alagamentos
custo_total = k * custo_reparo

# Exibindo os resultados
print(f"Probabilidade de ocorrer exatamente {k} alagamentos em um m√™s: {probabilidade_k_alagamentos:.4f}")
print(f"Custo total associado a {k} alagamentos: R$ {custo_total:.2f}")
```

Interpretamos isso no contexto do clima de S√£o Paulo, discutindo como essa rela√ß√£o afeta os eventos de alagamento e poss√≠veis estrat√©gias de mitiga√ß√£o.

```python
import numpy as np
import pandas as pd

# Dados hipot√©ticos para temperatura (em ¬∞C) e quantidade de chuva (em mm)
temperaturas = np.random.normal(loc=25, scale=4, size=365)  # 365 dias de temperatura
chuvas = np.random.normal(loc=20, scale=5, size=120)  # 120 dias de chuva

# Criando um DataFrame
dados = pd.DataFrame({
    'Temperatura': temperaturas,
    'Chuvas': np.concatenate([chuvas, np.zeros(245)])  # 245 dias sem chuva
})

# Calculando a correla√ß√£o
correlacao = dados['Temperatura'].corr(dados['Chuvas'])

# Exibindo os resultados
print(f"Coeficiente de correla√ß√£o entre temperatura e quantidade de chuva: {correlacao:.4f}")
```

**Gr√°ficos adicionais**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√¢metros da distribui√ß√£o normal para quantidade de chuva
mu_chuva = 20
sigma_chuva = 5

# Par√¢metros da distribui√ß√£o normal para temperatura
mu_temp = 25
sigma_temp = 4

# Valores para o eixo x
x_chuva = np.linspace(mu_chuva - 4*sigma_chuva, mu_chuva + 4*sigma_chuva, 1000)
x_temp = np.random.normal(mu_temp, sigma_temp, 120) # 120 dias chuvosos

# Fun√ß√£o densidade de probabilidade para a quantidade de chuva
pdf_chuva = norm.pdf(x_chuva, mu_chuva, sigma_chuva)

# Gr√°fico 1: Distribui√ß√£o da Quantidade de Chuva
plt.figure(figsize=(10,6))
plt.plot(x_chuva, pdf_chuva, label='Distribui√ß√£o Normal da Chuva', color='b')
plt.fill_between(x_chuva, pdf_chuva, where=(x_chuva >= 15) & (x_chuva <= 25), color='green', alpha=0.3, label='√Årea entre 15 e 25 mm')
plt.title('Distribui√ß√£o da Quantidade de Chuva em Dias Chuvosos')
plt.xlabel('Quantidade de Chuva (mm)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()

# Gerar dados para correla√ß√£o
np.random.seed(0)
chuva = np.random.normal(mu_chuva, sigma_chuva, 120)
correlacao = -0.5
temp = mu_temp + correlacao * (chuva - mu_chuva) + np.random.normal(0, sigma_temp, 120)

# Gr√°fico 2: Correla√ß√£o entre Temperatura e Chuva
plt.figure(figsize=(10,6))
plt.scatter(temp, chuva, color='purple', label=f'Correla√ß√£o: {correlacao}')
plt.title('Correla√ß√£o entre Temperatura e Quantidade de Chuva')
plt.xlabel('Temperatura M√©dia Di√°ria (¬∞C)')
plt.ylabel('Quantidade de Chuva (mm)')
plt.legend()
plt.grid(True)
plt.show()
```
