Gráfico sugerido para ilustrar a distribuição normal:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definindo a média e o desvio padrão
mu = 20  # média em mm
sigma = 5  # desvio padrão em mm

# Criando um intervalo de valores para a quantidade de chuva
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
# Calculando a função de densidade de probabilidade (PDF)
pdf = norm.pdf(x, mu, sigma)

# Plotando a distribuição normal
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribuição Normal', color='blue')
plt.title('Distribuição Normal da Quantidade de Chuva em Dias Chuvosos')
plt.xlabel('Quantidade de Chuva (mm)')
plt.ylabel('Densidade de Probabilidade')
plt.axvline(mu, color='red', linestyle='--', label='Média (20 mm)')
plt.legend()
plt.grid()
plt.show()

# Exemplo: calcular a probabilidade de um dia ter menos de 25 mm de chuva
prob_menos_25 = norm.cdf(25, mu, sigma)
print(f"Probabilidade de um dia chuvoso ter menos de 25 mm de chuva: {prob_menos_25:.4f}")
```

3. Probabilidade Condicional: Relação entre Temperatura e Chuva
Aplicamos o conceito de probabilidade condicional para analisar como a temperatura influencia a quantidade de chuva em um dia. Sabemos que a correlação (
𝜌
ρ) entre a quantidade de chuva e a temperatura é de -0.5, indicando uma correlação negativa.

A fórmula da distribuição conjunta normal bivariada foi utilizada para calcular a probabilidade condicional entre a chuva e a temperatura. (Aqui, descrevemos o cálculo que você fez anteriormente).

```python
from scipy.stats import norm

# Definindo os parâmetros para a temperatura
mu_T = 25  # média em °C
sigma_T = 4  # desvio padrão em °C

# Definindo o limite da temperatura
limite_temperatura = 24

# Calculando a probabilidade da temperatura ser menor que 24°C
P_T_menor_24 = norm.cdf(limite_temperatura, mu_T, sigma_T)

# Exibindo a probabilidade
print(f"Probabilidade de a temperatura ser menor que {limite_temperatura}°C: {P_T_menor_24:.4f}")
```

4. Probabilidade de Alagamentos e Custo Associado
Analisamos a frequência dos alagamentos nos dias chuvosos e o impacto financeiro deles. Sabemos que cada alagamento tem um custo médio de R$ 10.000.

Se a probabilidade de alagamentos nos dias chuvosos for calculada como
𝑃
(

)
P(Alagamento∣Chuva), podemos estimar os custos médios mensais associados ao alagamento.

(Descrever o cálculo que fizemos e seus resultados).

```python
from scipy.stats import poisson

# Definindo a média de alagamentos por mês
lambda_alagamentos = 10  # média de alagamentos
custo_reparo = 10000  # custo por alagamento

# Número de alagamentos que queremos calcular a probabilidade
k = 10

# Calculando a probabilidade de ocorrer exatamente k alagamentos
probabilidade_k_alagamentos = poisson.pmf(k, lambda_alagamentos)

# Calculando o custo total se ocorrerem k alagamentos
custo_total = k * custo_reparo

# Exibindo os resultados
print(f"Probabilidade de ocorrer exatamente {k} alagamentos em um mês: {probabilidade_k_alagamentos:.4f}")
print(f"Custo total associado a {k} alagamentos: R$ {custo_total:.2f}")
```

Interpretamos isso no contexto do clima de São Paulo, discutindo como essa relação afeta os eventos de alagamento e possíveis estratégias de mitigação.

```python
import numpy as np
import pandas as pd

# Dados hipotéticos para temperatura (em °C) e quantidade de chuva (em mm)
temperaturas = np.random.normal(loc=25, scale=4, size=365)  # 365 dias de temperatura
chuvas = np.random.normal(loc=20, scale=5, size=120)  # 120 dias de chuva

# Criando um DataFrame
dados = pd.DataFrame({
    'Temperatura': temperaturas,
    'Chuvas': np.concatenate([chuvas, np.zeros(245)])  # 245 dias sem chuva
})

# Calculando a correlação
correlacao = dados['Temperatura'].corr(dados['Chuvas'])

# Exibindo os resultados
print(f"Coeficiente de correlação entre temperatura e quantidade de chuva: {correlacao:.4f}")
```

**Gráficos adicionais**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros da distribuição normal para quantidade de chuva
mu_chuva = 20
sigma_chuva = 5

# Parâmetros da distribuição normal para temperatura
mu_temp = 25
sigma_temp = 4

# Valores para o eixo x
x_chuva = np.linspace(mu_chuva - 4*sigma_chuva, mu_chuva + 4*sigma_chuva, 1000)
x_temp = np.random.normal(mu_temp, sigma_temp, 120) # 120 dias chuvosos

# Função densidade de probabilidade para a quantidade de chuva
pdf_chuva = norm.pdf(x_chuva, mu_chuva, sigma_chuva)

# Gráfico 1: Distribuição da Quantidade de Chuva
plt.figure(figsize=(10,6))
plt.plot(x_chuva, pdf_chuva, label='Distribuição Normal da Chuva', color='b')
plt.fill_between(x_chuva, pdf_chuva, where=(x_chuva >= 15) & (x_chuva <= 25), color='green', alpha=0.3, label='Área entre 15 e 25 mm')
plt.title('Distribuição da Quantidade de Chuva em Dias Chuvosos')
plt.xlabel('Quantidade de Chuva (mm)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()

# Gerar dados para correlação
np.random.seed(0)
chuva = np.random.normal(mu_chuva, sigma_chuva, 120)
correlacao = -0.5
temp = mu_temp + correlacao * (chuva - mu_chuva) + np.random.normal(0, sigma_temp, 120)

# Gráfico 2: Correlação entre Temperatura e Chuva
plt.figure(figsize=(10,6))
plt.scatter(temp, chuva, color='purple', label=f'Correlação: {correlacao}')
plt.title('Correlação entre Temperatura e Quantidade de Chuva')
plt.xlabel('Temperatura Média Diária (°C)')
plt.ylabel('Quantidade de Chuva (mm)')
plt.legend()
plt.grid(True)
plt.show()
```
